```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
```

```python
df = pd.read_csv('Vfirm_sample.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().any(axis=0)
```

```python
dropna_df = df.dropna()
```

```python
dropna_df.describe()
```

```python
dropna_df['year2'] = dropna_df['year'].apply(lambda x: x ** 2)
```

```python
dropna_df.head()
```

```python
dropna_df['year3'] = dropna_df['year'].apply(lambda x: x ** 3)
```

```python
dropna_df['year4'] = dropna_df['year'].apply(lambda x: x ** 4)
```

```python
dropna_df.head()
```

```python
dropna_df['total_pay'].apply(np.log)
```

```python
dropna_df['total_pay2'] = dropna_df['total_pay'].apply(lambda x: x / 10000 )
```

```python
dropna_df.head()
```

```python
dropna_df['lsalary'] = dropna_df['salary'].apply(np.log)
```

```python
dropna_df.head()
```

```python
dropna_df['l_total_pay2'] = dropna_df.loc[:, 'total_pay2'].apply(np.log)
```

```python
dropna_df.head()
```

```python
dropna_df['l_total_hours'] = dropna_df.loc[:, 'total_hours'].apply(np.log)
```

```python
dropna_df['tenure2'] = dropna_df['tenure'].apply(np.square)
```

```python
dropna_df.head()
```

```python
year_dummies = pd.get_dummies(dropna_df['year'])
female_dummies = pd.get_dummies(dropna_df['sex'])
job_type_dummies = pd.get_dummies(dropna_df['job_type'])
edu_dummies = pd.get_dummies(dropna_df['edu'])
```

```python
year_dummies.head()
```

```python
year_dummies.columns = ['D2001', 'D2002', 'D2003', 'D2004', 'D2005', 'D2006']
```

```python
female_dummies.head()
```

```python
female_dummies.columns = ['male', 'female']
```

```python
female_dummies.head()
```

```python
job_type_dummies.head()
```

```python
job_type_dummies.columns = ['Djob_1', 'Djob_2', 'Djob_3', 'Djob_4']
```

```python
edu_dummies.head()
```

```python
edu_dummies.columns = ['Dedu_1', 'Dedu_2', 'Dedu_3', 'Dedu_4']
```

```python
dummies = pd.concat([year_dummies, female_dummies, job_type_dummies, edu_dummies], axis=1)
```

```python
dummies.head()
```

```python
ols_df = pd.concat([dropna_df, dummies], axis=1)
ols_df.head()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
ols_df.columns
```

```python
ols_df['age2'] = ols_df['age'].apply(np.square)
ols_df['age3'] = ols_df['age'].apply(lambda x: x ** 3)
ols_df['age4'] = ols_df['age'].apply(lambda x: x ** 4)
```

```python
ols_df.head()
```

```python
ols_df.columns
```

```python
formula = 'l_total_pay2 ~ age + age2 + age3 + age4 + tenure + tenure2 + female + Dedu_2 + Dedu_3 + Dedu_4 + Djob_1 + Djob_1 + Djob_3 + Djob_4 + D2002 + D2003 + D2004 + D2005 + D2006 + l_total_hours'
```

```python
results = smf.ols(formula, ols_df).fit()
print(results.summary())
```
